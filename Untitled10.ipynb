```python
import requests
from PIL import Image
from io import BytesIO
import glob
from ultralytics import YOLO
import matplotlib.pyplot as plt

# List of image URLs
image_urls = [
    'https://th.bing.com/th/id/R.cf1204e7d0e5eac3310137a8a924e2d1?rik=N%2bKRRUxpDD23aw&riu=http%3a%2f%2f1.bp.blogspot.com%2f-VNJdCO_BnBg%2fUK4XVLJh91I%2fAAAAAAAAAxg%2fWjzC76MIUo4%2fs1600%2f38.jpg&ehk=3bwb%2f%2fWNywM%2fag8F3NHlSGkkUBII9OUuo0OZcBe%2bW8E%3d&risl=&pid=ImgRaw&r=0'
]

# Download and save images
for i, url in enumerate(image_urls):
    response = requests.get(url, stream=True) # Changed to handle the image stream
    img = Image.open(response.raw) # Open the image from the stream
    img.save(f'image_{i}.jpg')


model = YOLO('yolov8n.pt')

# List of downloaded image paths
image_paths = glob.glob('image_*.jpg')

# Process each image
for image_path in image_paths:
    # Perform detection
    results = model(image_path)

    # results is a list, so we iterate through each result
    for result in results:
        # Convert to PIL Image
        result_image = Image.fromarray(result.plot())

        # Show the result
        plt.imshow(result_image)
        plt.axis('off')
        plt.show()
```
