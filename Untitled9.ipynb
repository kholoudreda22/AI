```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from matplotlib import pylab as plt

import pandas as pd
```

```python
def find_boundary(img):
    rows = np.any(img, axis=1)
    cols = np.any(img, axis=0)

    ymin, ymax = np.where(rows)[0][[0, -1]]
    xmin, xmax = np.where(cols)[0][[0, -1]]

    ymin, ymax = ymin / 28.0, ymax / 28.0
    xmin, xmax = xmin / 28.0, xmax / 28.0
    width = xmax - xmin
    height = ymax - ymin

    return [xmin, ymin, width, height]

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# Apply the boundary function to each image in the dataset
y_train_boundary = np.array([find_boundary(img) for img in X_train])
y_test_boundary = np.array([find_boundary(img) for img in X_test])
```

```python
# Define the model with two output layers
input_layer = Input(shape=(28, 28, 1))
x = Conv2D(32, (3, 3), activation='relu')(input_layer)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu')(x)
x = MaxPooling2D((2, 2))(x)
x = Flatten()(x)
x = Dense(64, activation='relu')(x)

# Classification output
classification_output = Dense(10, activation='softmax', name='classification')(x) # 10 outputs for digits 0-9

# Localization output
localization_output = Dense(4, name='localization')(x)

model = Model(inputs=input_layer, outputs=[classification_output, localization_output])
model.compile(optimizer='adam',
              loss={'classification': 'sparse_categorical_crossentropy',
                    'localization': 'mean_squared_error'})
model.summary()

# Training
history = model.fit(
    X_train,
    {'classification': y_train, 'localization': y_train_boundary},
    epochs=10,
    batch_size=32,
    validation_data=(X_test, {'classification': y_test, 'localization': y_test_boundary})
)
```

```python
import matplotlib.pyplot as plt

def display_predictions(model, X, y_true, num_samples):
    y_pred , _ = model.predict(X)
    for i in range(num_samples):
        img = X[i].reshape(28, 28)
        true_box = y_true[i]
        pred_box = y_pred[i]
        fig, ax = plt.subplots(1)
        ax.imshow(img, cmap='gray')
        rect_true = plt.Rectangle(
            (true_box[0] * 28, true_box[1] * 28),
            true_box[2] * 28, true_box[3] * 28,
            linewidth=2, edgecolor='g', facecolor='none'
        )
        ax.add_patch(rect_true)
        rect_pred = plt.Rectangle(
            (pred_box[0] * 28, pred_box[1] * 28),
            pred_box[2] * 28, pred_box[3] * 28,
            linewidth=2, edgecolor='r', facecolor='none'
        )
        ax.add_patch(rect_pred)
        plt.show()
display_predictions(model, X_test, y_test_boundary, 5)
```
