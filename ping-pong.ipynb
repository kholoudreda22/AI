```python
import pygame
pygame.init()
width , length = 700,500
winner = pygame.display.set_mode((width,length))
pygame.display.set_caption("Ping Pong")
timetoscreenUpdate = 60
White = (255,255,255)
Black = (0,0,0)
paddle_width,paddle_height = 20,100
radiusofball = 7
scoreFont = pygame.font.SysFont(" line " , 50)
scoretoWin = 10
class Paddle :
  color= White
  v = 4
  def init(self,x,y,width,height):
    self.x = original_x =x
    self.y = original_y =y
    self.width = width
    self.height = height
    def draw_paddle (self,win) :
      pygame.rect(win , self.color , (self.x,self.y,self.width , self.height))

    def move (self,up = True ) :
      if up == True :
        self.y -= self.v
      else:
        self.y += self.v
class Ball:
     maxValue = 5
     color = White
     def ReturnPaddletoitsplace (self):
        self.x=self.original_x
        self.y=self.original_y
        self.y=self.original_y
        self.radius=radiusofball
        self.x_value = self.maxValue
        self.y_value = 0
     def draw_paddle (self , win ):
        pygame.draw.circle(win,self,self.color , (self.x,self.y),self.radius)
#to move x and y indepentaly
     def move (self) :
          self.x += self.v
          self.y += self.v

     def ReturnPaddletoitsplace (self):
        self.x=self.original_x
        self.y=self.original_y
        self.y_value = 0
        self.x *= -1 #to reverse the direction
     def draw (win,paddles , ball , left_score , right_score ):
         win.fill("black")
         left_score_text = scoreFont.render(f"{left_score}",1,White)
         right_score_text = scoreFont.render(f"{right_score}",1,White)
         win.blit (right_score_text , (width//4-left_score_text.get_width()//2,20))
         win.blit (right_score_text (width* 3/4-right_score_text.get_width()%2,20))
         for paddle in paddles :
           paddle.draw(win)
         for i in range(10,paddle_height,paddle_height//20):
          if i % 2 == 1:
           continue
         pygame.draw.rect(win , White , width//2-5,paddle_height//20)

         ball.draw(win)
pygame.display.update()

def collision (ball,left_paddle , right_paddle ) : #if we have collisions
  if ball.y + ball.radius >= paddle_height :
    ball.y_value *= -1
  elif ball.y - ball.y_radius :

   if ball.x_value < 0:
     if ball.y >= left_paddle.y and ball.y <= left_paddle.y + left_paddle.paddle.height :
       if ball.x - ball.radius <= left_paddle.x + left_paddle.paddle.width :
         ball.x_value *= -1
         middle_y = left_paddle + left_paddle.paddle.paddle_height
         differencY = middle_y - ball.y
         reductionF = (left_paddle.height/2)/ball.maxValue
         y_value = differencY / reductionF
         ball.y_value = -1 * y_value
  else :
    if ball.y>= right_paddle.y and ball.y <= right_paddle.y :
      if ball.x + ball.radius >= right_paddle.x :
        ball.x_value *= -1
        middle_y= right_paddle.y + right_paddle.height/2
        differencY = middle_y - ball.y
        reductionF = (right_paddle.height/2)/ball.maxValue
        ball.y_value = -1 * y_value

def howPaddleMove (keys,left_paddle , right_paddle) :
    if keys[pygame.K_w] and left_paddle.y - left_paddle.v >=0:
      left_paddle.move(up=True)
    if keys[pygame.k_w] and left_paddle.y + left_paddle.v +left_paddle.height <= length :
      left_paddle.move(up=False)
    if keys[pygame.K_UP] and right_paddle.y - left_paddle.v >=0 :
      right_paddle.move(up=True)
      if keys[pygame.K_DOWN] and right_paddle.y + right_paddle.v +right_paddle.length <= length :
       right_paddle.move(up=False)
def main ():
     run=True
     clock=pygame.time.clock()
     left_paddle=Paddle(10,length/2 - paddle_height//2 , paddle_width , paddle_height)
     right_paddle=Paddle(width -10 - paddle_width , length//2 - paddle_height//2 , paddle_width , paddle_height )
     ball = Ball(width//2 , length//2 , Ball.radius)
     left_score = 0
     right_score =0
     while run :
       clock.tick(timetoscreenUpdate)
       draw(win,[left_paddle,right_paddle],ball,left_score,right_score)

       for event in pygame.event.get ():
         if event.type == pygame.QUIT:
           run=False
           break
       keys = pygame.key.get_pressed ()
       howPaddleMove(keys,left_paddle,right_paddle)
       ball.move()
       collision(ball , left_paddle ,right_paddle)

       if ball.x < 0 :
         right_score +=1
         ball.ReturnPaddletoitsplace
       elif ball.x > width :
         left_score += 1

         won = False
         if left_score >= winner :
           won=True
           win_text = "left player won"
         elif right_score >= winner :
           won=True
           win_text = "Right player won"
         if won == True :
           text = scoreFont.render(win_text,1,White)
           win.blit(text,(width//2 - text.get_width()//2 , length//2 - text.get_height()//2))
           pygame.display.update()
           pygame.time.delay(5000)
           ball.ReturnPaddletoitsplace()
           right_paddle.ReturnPaddletoitsplace ()
           left_paddle.ReturnPaddletoitsplace ()
           left_score =0
           right_score=0
     pygame.quit()
if __name__== 'main' :
       main()
```
